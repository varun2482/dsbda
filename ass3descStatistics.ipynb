```python
def count_values(series):
  return(series.count())
def max_value(series):
  temp = series[0]
  for i in series:
   if i > temp:
     temp = i
  return temp
def min_value(series):
  temp = series[0]
  for i in series:
   if i < temp:
    temp = i
  return temp
def mean_value(series):
    sum = 0
    for i in series:
        sum = i + sum
    return round(sum/count_values(series),6)
def median_value(series):
    count = count_values(series)
    sorted_series = sorted(series)
    if count%2 != 0:
        return sorted_series[(count//2)-1]
    else:
        return (sorted_series[(count//2)-1] + sorted_series[count//2])/2
def quantile_values(series, p):
    sorted_series = sorted(series)
    R1 = p/100 *(count_values(series)+1)
    if R1%10 == 0:
        return(sorted_series[R1])
    else:
        R1 = int(R1)
        R2 = R1+1
        return( ((p/100)*(sorted_series[R2] - sorted_series[R1])) + sorted_series[R1])
def std_values(series):
    mean = mean_value(series)
    sum = 0
    series = ((series.values).tolist())
    for i in range(len(series)):
        sum = sum + np.square(series[i] - mean)
    return round(np.sqrt(sum/len(series)),6)
df.info()
```

```python
columns = ['age', 'fnlwgt', 'educational-num', 'capital-gain',
'capital-loss', 'hours-per-week']
for i in columns:
  print("VALUES FOR COLUMN -->", i)
  print("Count\t",count_values(df[i]))
  print("Max\t",max_value(df[i]))
  print("Min\t",min_value(df[i]))
  print("Mean\t",mean_value(df[i]))
  print("Median\t",median_value(df[i]))
  print("Std\t",std_values(df[i]))
  print("25%\t",quantile_values(df[i], 25))
  print("50%\t",quantile_values(df[i], 50))
  print("75%\t",quantile_values(df[i], 75))
  print("\n-------------------------------------------------------------------------\n")
```
