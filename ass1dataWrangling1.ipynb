```python
df['Date'] = df['Date'].str.replace("'", "").str.replace('/', '')
```

```python
df['Year'] = df['Date'].str[:4]
df['Month'] = df['Date'].str[4:6]
df['Day'] = df['Date'].str[6:8]
```

```python
df['Date'] = pd.to_datetime(df['Year'] + df['Month'] + df['Day'], format="%Y%m%d")
df = df.drop(['Year', 'Month', 'Day'], axis=1)
df
```
