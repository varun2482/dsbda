```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df.head()
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='sex',y='age')
```

```python
Q1_male=df[df['sex']=='male']['age'].quantile(.25)
Q3_male=df[df['sex']=='male']['age'].quantile(.75)

print(Q1_male)
print(Q3_male)
```

```python
Q1_female=df[df['sex']=='female']['age'].quantile(.25)
Q3_female=df[df['sex']=='female']['age'].quantile(.75)

print(Q1_female)
print(Q3_female)
```

```python
Q1_male_survived=df[((df['sex']=='male') & (df['survived']==1))]['age'].quantile(.25)
Q3_male_survived=df[((df['sex']=='male') & (df['survived']==1))]['age'].quantile(.75)

print(Q1_male_survived)
print(Q3_male_survived)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='age',y='sex')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='sex',y='age',hue='survived')
```
