```python
a='once upon a time, there was a Greek King, Midas. He was very rich and had lots of Gold. He had a daughter, who he loved a lot.One day, Midas found an angel in need of help. He helped her and in return she agreed to grant a wish.'
```

```python
from nltk.tokenize import sent_tokenize
a=a.lower()
print(a)
```

```python
import nltk
nltk.download('punkt')
```

```python
b=sent_tokenize(a)
b
```

```python
from nltk.tokenize import word_tokenize
```

```python
c=word_tokenize(a)
c
```

```python
from nltk.stem import PorterStemmer
porter=PorterStemmer()
stemmed_word=[porter.stem(word) for word in c]
stemmed_word
```

```python
import nltk
nltk.download('wordnet')

import nltk
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
word_lemmatizer=[lemmatizer.lemmatize(word) for word in stemmed_word]
word_lemmatizer
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag
pos_var=pos_tag(word_lemmatizer)
for word_lemmatizer,pos_tag in pos_var:
  print(f"{word_lemmatizer}:{pos_tag}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
d0="good boy"
d1="girl good"
d2="boy girl good"
tfidf=TfidfVectorizer()
string=[d0,d1,d2]
string
```

```python
result=tfidf.fit_transform(string)
print("tfidf values:|")
print(result)
```

```python
for feature_name, idf_value in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(feature_name, ":", idf_value)
```

```python
print("\nWord Indexes:")
print(tfidf.vocabulary_)
```
