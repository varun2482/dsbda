```python
Q1=df['Attendance'].quantile(0.25)
Q3=df['Attendance'].quantile(0.75)

print(Q1,'\n',Q3)
```

```python
IQR=Q3-Q1
ul=Q3+1.5*IQR
ll=Q1-1.5*IQR
print(ul,'\n',ll)
```

```python
df[(df['Attendance']>ul) | (df['Attendance']<ll)]
```

```python
new_df=df[(df['Attendance']<ul) & (df['Attendance']>ll)]
new_df
```

```python
def remove_outlier(val):
  if(val>ul):
    return(ul)
  elif(val<ll):
    return(ll)
  else:
    return(val)

df['Attendance']=df['Attendance'].apply(remove_outlier)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(data=df['Attendance'])

plt.subplot(1,2,2)
sns.boxplot(data=df['Attendance'],orient='y')
```
