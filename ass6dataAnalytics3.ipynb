```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
data=sns.load_dataset('iris')
```

```python
x=data.iloc[:,:4]
type(x)
```

```python
y=data['species']
type(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
x_train.shape
```

```python
x_train.shape
```

```python
type(x_train)
```

```python
type(y_train)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_test = sc.fit_transform(x_test)
x_train = sc.fit_transform(x_train)
```

```python
type(x_test)
```

```python
x_test
```

```python
type(x_train)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()
```

```python
type(nb_model)
nb_model.fit(x_train,y_train)
```

```python
y_predict=nb_model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score is :", accuracy_score(y_test,y_predict))
```

```python
from sklearn.metrics import precision_score
print("Precision score is :", precision_score(y_test,y_predict,average='macro'))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix is :", confusion_matrix(y_test,y_predict))
```

```python
from sklearn.metrics import classification_report
print("Classification report is :", classification_report(y_test,y_predict))
```
